```python
import pandas as pd
data = {
    "movie_id": [101,102,103,104,105],
    "title": ["Dhurandar","Life of Pie","Mahavatar Narasimha", "Kantara Chapter 1", "F1"],
    "Genre": ["Action","Fiction", "Mythology", "Drama", "Action"],
    "Release_Year": [2025, 2013, 2025, 2025, 2025],
    "Rating" : [9.4,8,10,10,9.5],
    "Box_office": ["2 crore", "90 lakhs", "100 crore", "200 crore", "20 crore"],
    "Duration_mins": [250,200,205,216,200],
    "Platform": ["theatre","theatre","theatre","theatre","theatre"]
}
df = pd.DataFrame(data)

def convert_to_crore(value):
  if "crore" in value:
    return float(value.replace(" crore", ""))
  elif "lakhs" in value:
    return float(value.replace(" lakhs", ""))/100
  return value

df["Box_office"] = df["Box_office"].apply(convert_to_crore)
df
```
